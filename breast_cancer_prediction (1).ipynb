```python
df['diagnosis'] = df['diagnosis'].map ({'M':1,'B':0})
df['diagnosis'].unique()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop('diagnosis',axis=1)
y = df['diagnosis']
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

▶
**feature scaling:**
Our dataset may contains features in varying magnitude, units, and range. so we have to scale these variations in one form.



```python
from sklearn.preprocessing import StandardScaler 
ss = StandardScaler()
x_train = ss.fit_transform (x_train)
x_test = ss.fit_transform (x_test)
```

**Model Selection:**
we have cleaned our data and now we have to choose one machine learning algorithm which gives the best accuracy in final result. here the output of our data is categorical format so we'll use the supervised classification machine learning algorithms. so for the best accuracy, we have used the various machine learning algorithm model so that we find out the best model for detecting the breast cancer. so we have 4 following models: 

*   Logistic Rgression
*   Decision Tree Classifier

*   Random Forest Classifier
*   Support Vector Classifier








**▪ Logistic Regression**



```python
# importing the libraries for logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
from sklearn.metrics import precision_score,recall_score, f1_score
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
x_train
```

```python
y_train
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
y_test
```

```python
# Accuracy Score for the logistic regression
print("Accuracy Score of Logistic Regression:")
acc1 = accuracy_score(y_test, y_pred)
print(acc1)    
```

```python
print("Logistic Regression Accuracy, Presicion, Recall and F1_Score are : ")
pre1 = precision_score(y_test, y_pred)
print("precision score : ",pre1)
rec1 = recall_score(y_test, y_pred)
print("Recall Score : ",rec1)
f1 = f1_score(y_test, y_pred)
print("F1 score : ",f1)
print("Accuracy Score : ", acc1)
```

```python
print("Classifiaction Report for Logistic Regression :\n")
print(classification_report(y_test,y_pred))
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred),annot = True)
plt.show()
```
